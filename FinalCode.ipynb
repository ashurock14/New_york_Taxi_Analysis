```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['RatecodeID'].value_counts()
```

```python
df.index.plot(kind='scatter',y='total_amount',x='index',figsize=(10,5))
```

```python
df=df.reset_index()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(data=df,x='index',y='total_amount',hue='trip_distance')
```

```python
less_than_1000=df[df['total_amount']<1000].reset_index()
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(data=less_than_1000,x='index',y='total_amount')
```

```python
negative_amount=df[df['total_amount']<0].reset_index()
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.scatterplot(data=negative_amount,x='index',y='total_amount')
```
